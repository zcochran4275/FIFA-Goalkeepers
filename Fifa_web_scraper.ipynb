```python
from bs4 import BeautifulSoup
import requests
import csv 

csv_file = open('Fifa GK Stats.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Name', 'League', 'Age', 'Height(cm)', 'Weight(kg)', 'OVR', 'Potential', 'Value', 'Wage', 'Minutes Played', 'Appearances', 'Saves', 'Saves Inside the Box', 'Penalties Save', 'Clean Sheets', 'League Rating', 'International Reputation', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions','Balance', 'Jumping', 'Stamina', 'Strength', 'Aggression', 'Interceptions', 'Vision', 'Penalties', 'Composure', 'Diving', 'Handling', 'Kicking', 'Positioning', 'Reflexes', 'Link'])
links = []
for i in range(0,2160,60):
    source = requests.get('https://sofifa.com/players?pn=0&offset={}'.format(str(i))).text

    soup = BeautifulSoup(source, 'lxml')

    name_link = {}

    col_names = soup.find_all('td', class_='col-name')

    for col_name in col_names:
        col_name = col_name.a
        link = "https://sofifa.com" + str(col_name).split('href="')[1].split("\"")[0]
        if link.split('/')[3] == 'player':
            links.append(link)

for link_ in links:
    player_source = requests.get(link_).text
    
    p_soup = BeautifulSoup(player_source, 'lxml')
    
    grid = p_soup.find("div", class_ = 'grid')
    

    p_info = grid.find("div", class_ = 'info').text.split("\n")

    p_name = link_.split('/')[5].split('-')
    p_name = " ".join([i.capitalize() for i in p_name])
    p_age = p_info[1].split("K ")[1].split('y')[0]
    p_height = p_info[1].split(') ')[1].split(" ")[0].split("cm")[0]
    p_weight = p_info[1].split(') ')[1].split(" ")[1].split('kg')[0]

    ovrs_and_value = grid.find("section", class_ = "card spacing")
    p_ovr = ovrs_and_value.find_all('span', class_ = 'bp3-tag')[0].text
    p_potential = ovrs_and_value.find_all('span', class_ = 'bp3-tag')[1].text
    p_value = ovrs_and_value.find_all('div', class_ = 'block-quarter')[2].text.split('V')[0][1:]   #come back and format these numbers
    p_wage = ovrs_and_value.find_all('div', class_ = 'block-quarter')[3].text.split('W')[0][1:] 

    p_profile = grid.find('ul', class_ = 'pl').find_all('li', class_ = 'ellipsis')
    p_int_rep = p_profile[3].text.split('  I')[0]
    p_workrate = p_profile[4].text.split('Rate')[1]

    p_stats = p_soup.find_all('div',class_ ='col col-12')[1].find_all('ul', class_ = 'pl')

    p_movement = p_stats[2].find_all('li')
    p_acceleration = p_movement[0].text.split(' ')[0][0:2]
    p_sprint_speed = p_movement[1].text.split(' ')[0][0:2]
    p_agility = p_movement[2].text.split(' ')[0][0:2]
    p_reactions = p_movement[3].text.split(' ')[0][0:2]
    p_balance = p_movement[4].text.split(' ')[0][0:2]

    p_power = p_stats[3].find_all('li')
    p_jumping = p_power[1].text.split(' ')[0][0:2]
    p_stamina = p_power[2].text.split(' ')[0][0:2]
    p_strength = p_power[3].text.split(' ')[0][0:2]

    p_mentality = p_stats[4].find_all('li')
    p_aggression = p_mentality[0].text.split(' ')[0][0:2]
    p_interceptions = p_mentality[1].text.split(' ')[0][0:2]
    p_vision = p_mentality[3].text.split(' ')[0][0:2]
    p_penalties = p_mentality[4].text.split(' ')[0][0:2]
    p_composure = p_mentality[5].text.split(' ')[0][0:2]

    p_goalkeeping = p_stats[6].find_all('li')
    p_diving = p_goalkeeping[0].text.split(' ')[0][0:2]
    p_handling = p_goalkeeping[1].text.split(' ')[0][0:2]
    p_kicking = p_goalkeeping[2].text.split(' ')[0][0:2]
    p_positioning = p_goalkeeping[3].text.split(' ')[0][0:2]
    p_reflexes = p_goalkeeping[4].text.split(' ')[0][0:2]

    link_extra = "/".join(link_.split('/')[0:6])+ '/live'
    player_source_ = requests.get(link_extra).text
    p_soup_ = BeautifulSoup(player_source_, 'lxml')
    print(link_extra)
    try:
        p_table = p_soup_.find('tbody')
        p_table_stats = p_table.find_all('tr')[1]
        p_table_stats = p_table_stats.find_all('td')

        p_league  = p_table_stats[2].text
        p_min = p_table_stats[3].text
        p_appearance = p_table_stats[4].text
        p_saves = p_table_stats[39].text
        p_saves_inbox = p_table_stats[40].text
        p_penalties_saved = p_table_stats[41].text
        p_clean_sheets = p_table_stats[42].text
        p_rating = p_table_stats[45].text
    except:
        pass


    csv_writer.writerow([p_name, p_league, p_age, p_height, p_weight, p_ovr,p_potential,p_value,p_wage,p_min, p_appearance, p_saves, p_saves_inbox, p_penalties_saved, p_clean_sheets, p_rating, p_int_rep,p_acceleration,p_sprint_speed,p_agility,p_reactions,p_balance,p_jumping,p_stamina,p_strength,p_aggression,p_interceptions,p_vision,p_penalties,p_composure,p_diving,p_handling,p_kicking,p_positioning,p_reflexes,link_])









```
