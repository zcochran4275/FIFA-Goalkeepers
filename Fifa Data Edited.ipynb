```python
import pandas as pd
import numpy as np



fifa_data = pd.read_excel("Fifa GK Stats Raw.xlsx")

fifa_data["Appearances"].astype('int')

fifa_data["Appearances"] = [int(x) for x in fifa_data['Appearances']]
fifa_data["Saves"] = [int(x) for x in fifa_data['Saves']]
fifa_data["Minutes Played"] = [int(x) for x in fifa_data['Minutes Played']]
fifa_data["Saves Inside the Box"] = [int(x) for x in fifa_data['Saves Inside the Box']]
fifa_data["Penalties Save"] = [int(x) for x in fifa_data['Penalties Save']]
fifa_data["Clean Sheets"] = [int(x) for x in fifa_data['Clean Sheets']]
fifa_data["Value"] = [str(x) for x in fifa_data['Value']]
fifa_data["Wage"] = [str(x) for x in fifa_data['Wage']]
fifa_data["League"] = [str(x) for x in fifa_data['League']]

fifa_data["Value"]
values = []
for i in fifa_data["Value"]:
    if i[-1] == 'M':
        value = float(i[0:-1]) *1000000
        values.append(int(value))
    elif i[-1] == 'K':
        value = float(i[0:-1]) *1000
        values.append(int(value))
    else:
        try:
            value = i
            values.append(int(i))
        except:
            values.append(None)
fifa_data['Value'] = values



wages = []
for i in fifa_data["Wage"]:
    if i[-1] == 'M':
        wage = float(i[0:-1]) *1000000
        wages.append(int(wage))
    elif i[-1] == 'K':
        wage = float(i[0:-1]) *1000
        wages.append(int(wage))
    else:
        try:
            wage = i
            wages.append(int(i))
        except:
            wages.append(None)
fifa_data['Wage'] = wages

fifa_data["Projected Growth"] = fifa_data['Potential'] - fifa_data['OVR']

fifa_data = fifa_data.reindex(columns=['Name', 'League', 'Age', 'Height(cm)', 'Weight(kg)', 'OVR', 'Potential', 'Projected Growth', 'Value', 'Wage', 'Minutes Played', 'Appearances', 'Saves', 'Saves Inside the Box', 'Penalties Save', 'Clean Sheets', 'League Rating', 'International Reputation', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions','Balance', 'Jumping', 'Stamina', 'Strength', 'Aggression', 'Interceptions', 'Vision', 'Penalties', 'Composure', 'Diving', 'Handling', 'Kicking', 'Positioning', 'Reflexes', 'Link'])

fifa_data["Fitness"] = (fifa_data['Acceleration'] + fifa_data["Sprint Speed"] + fifa_data['Agility'] + fifa_data['Stamina'] + fifa_data['Balance'] + fifa_data['Strength']+ fifa_data['Jumping'])/7

fifa_data.head()


 #fifa_data.to_excel('Fifa Data.xlsx')



```
